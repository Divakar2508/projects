```python
from autoscraper import AutoScraper
import pandas as pd
```

```python
amazon_url="https://www.amazon.in/s?k=iphone+13"

wanted_list=["₹66,900","Apple iPhone 13 (128GB) - Midnight"," 4,464"]
```

```python
scraper=AutoScraper()
result=scraper.build(amazon_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(amazon_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_4wba':'Price','rule_2dqm':'Title','rule_q1eh':'Review'})
scraper.keep_rules(['rule_4wba','rule_2dqm','rule_q1eh'])
scraper.save('amazon-search')
```

```python
results=scraper.get_result_similar('https://www.amazon.in/s?k=poco',group_by_alias=True)
```

```python
results['Price']
```

```python
results['Review']
```

```python
df=pd.DataFrame(dict([(k,pd.Series(v)) for k,v in results.items() ]))
```

```python
df
```

```python
df=df.dropna()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df["Price"]= df["Price"].str.replace('₹','')
```

```python
df["Price"]= df["Price"].str.replace(',','')
```

```python
df['Price']=df['Price'].astype(int)
```

```python
df['Review']=df['Review'].astype(int)
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
plt.hist(df['Price'])
plt.title("Price")
plt.show()
```

```python
plt.figure(figsize=(27,10))
sns.barplot('Title','Price',data=df ,palette='gist_rainbow_r')
plt.tight_layout()
plt.xlabel('Title',fontsize=45)
plt.ylabel('Price',fontsize=45)
plt.xticks(rotation = 90,fontsize=40)
plt.yticks(fontsize=40)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(27,10))
sns.barplot('Title','Review',data=df ,palette='gist_rainbow_r')
plt.tight_layout()
plt.xlabel('Title',fontsize=45)
plt.ylabel('Review',fontsize=45)
plt.xticks(rotation = 90,fontsize=40)
plt.yticks(fontsize=40)
plt.legend()
plt.show()
```

```python
sns.distplot(df['Price'],color='darkred')
print("Maximum Price of the Mobile:",df['Price'].max())
sns.set(rc={'figure.figsize':(10,6)})
```

```python
sns.distplot(df['Review'],color='darkred')
print("Highest Review is :",df['Review'].max())
sns.set(rc={'figure.figsize':(10,6)})
```
